```python
#Mean Square Error(MSE)=Cost function 
#MSE=Sum of sqaure of delta(differece from actual value)  /  total number of elements
#Gradient descent is an algorithm that finds best fit line for given traing data set
#mse = [summation(y - (mx+b))^2]/n
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
def gradient_descent(x,y):
    m_curr=b_curr=0
    iterations=100
    n=len(x)
    learning_rate=0.01 #dont miss the global minima
    plt.scatter(x,y,color='red',marker='+',linewidth='5')
    for i in range(iterations):
        yp=m_curr*x + b_curr
        plt.plot(x,yp,color='green')
        cost=(1/n)*sum([val**2 for val in (y-yp)])
        #pde wrt m of mse
        md=-(2/n)*sum(x*(y-yp))
        #pde wrt b of mse
        bd=-(2/n)*sum(y-yp)
        m_curr=m_curr - learning_rate * md
        b_curr=b_curr - learning_rate * bd
     
        print("m{} ,b{} ,cost{}, iteration{}".format(m_curr, b_curr, cost, i))
        
    pass
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
